# **Graph #1**: Gradient is zero


```
from mpl_toolkits import mplot3d
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x, y):
    return (-x ** 2 - y ** 2 + 10)


# Create the data
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)
z = np.linspace(-10, 10, 1000)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)


# Plot the graph
fig = plt.figure()
plt.title("3D Projection")
ax = plt.axes(projection='3d')


# Point
ax = fig.add_subplot(111, projection='3d')
x =0
y = 0
z =10

# Show graph
ax.view_init(46, 60)   #el,az
ax.contour3D(X, Y, Z, 40, cmap='Greys')
ax.scatter(x, y, z, c='r', marker='o')

# Label axis 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


# Quiver 
```

The point (in red) is at the peak of the inverted hyperbola. The gradient at that point is 0. Point: (0,0,10)

### Quiver for Graph #1 

```
def dx(x):
    return (-2*x )
  
def dy(y):
    return ( - 2*y)


x = np.linspace(-20, 20, 10)
y = np.linspace(-20, 20, 10)
X, Y = np.meshgrid(x, y)

# Quiver 
U = dx(X)
V = dy(Y)

fig1, ax1 = plt.subplots()
#ax1.set_title('Arrows scale with plot width, not view')
Q = ax1.quiver(X, Y, U, V, units='width')
#qk = ax1.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
                   #coordinates='figure')

plt.show()
```

# **Graph #2**: Local Max 

```
#fig2 = plt.figure();
#ax = plt.axes(projection= '3d')

def f(x, y):
    return (x ** 2 /16 - y ** 2 /25)

# Create the data
x = np.linspace(-6, 6, 1000)
y = np.linspace(-6, 6, 1000)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the graph
fig = plt.figure()
ax = plt.axes(projection='3d')

# Point
ax = fig.add_subplot(111, projection='3d')
x = 0
y = 0
z = 0

# Show graph
ax.contour3D(X, Y, Z, 100, cmap='binary')
ax.scatter(x,y,z, c='r', marker='v')

# Label axis 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

The point (in red) is located at the saddle point, which is a local maximum. In this case it is not a global maximum. Point: (0,0,0)

# **Graph #3**: Local Min

```
def f(x, y):
    return (np.cos(x) + np.sin(y))

# Create data for the graph
x = np.linspace(-8, 8, 1000)
y = np.linspace(-8, 8, 1000)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the graph
fig = plt.figure()
ax = plt.axes(projection='3d')

# Point
#ax = fig.add_subplot(111, projection ='3d')
a = 3.5
b = 8
d = 0.2

# Show the graph
ax.view_init(30, 60)
ax.scatter(a,b,d, c='red', marker='o')
#ax.set_zlim(0, 2)
ax.contour3D(X, Y, Z, 100, cmap='binary')

# Label axis
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

The point (in red) is a local minimum, it is not a global minimum in this case. Point: (3.5,8,0)
